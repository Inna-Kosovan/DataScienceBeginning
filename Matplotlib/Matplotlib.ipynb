### Functional method

```python
plt.plot(a, b)
plt.xlabel('a array')
plt.ylabel('b array')
plt.title('Plot for a and b arrays')
```

```python
plt.subplot(1, 2, 1)
plt.plot(a, b, 'r')
plt.subplot(1, 2, 2)
plt.plot(b, a, 'b')
```

# Object-oriented method

```python
figure = plt.figure()
axes = figure.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(a,b)
axes.set_xlabel('a array values')
axes.set_ylabel('b array values')
axes.set_title('Some title')
```

```python
figure1 = plt.figure()
axes1 = figure1.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = figure1.add_axes([0.3, 0.6, 0.5, 0.4])
```

```python
figure1 = plt.figure()
axes1 = figure1.add_axes([0.1, 0.0, 0.8, 0.8])
axes2 = figure1.add_axes([0.4, 0.0, 0.5, 0.4])
```

```python
figure1 = plt.figure()
axes1 = figure1.add_axes([0.0, 0.0, 1, 1])
axes2 = figure1.add_axes([0.1, 0.4, 0.5, 0.5])

axes1.plot(a,b)
axes2.plot(b,a)

axes1.set_xlabel('a array values')
axes1.set_ylabel('b array values')
axes1.set_title('Axes1 title')

axes2.set_xlabel('b array values')
axes2.set_ylabel('a array values')
axes2.set_title('Axes2 title')
```
