```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
dots = sns.load_dataset('dots')
```

```python
dots.head()
```

```python
sns.displot(dots['firing_rate'], kde=True, bins=25)
```

```python
sns.jointplot(x='coherence', y='firing_rate', data=dots)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.displot(tips['total_bill'], kde=True, bins=25)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', color='tomato')
```

```python
sns.pairplot(tips, hue='sex', palette='husl', hue_order=['Female', 'Male'])
```

```python
sns.rugplot(tips['total_bill'])
```
